```python
import tensorflow as tf
```

```python
(train_dataset, validation_dataset) = tf.keras.preprocessing.image_dataset_from_directory(
    'dataset-procesado',
    validation_split=0.3,
    label_mode='int',
    subset='both',
    seed=42,
    image_size=(60, 60)
)

class_names = sorted(train_dataset.class_names)
```

```python
train_dataset = train_dataset.map(
    lambda x, y: (tf.keras.applications.mobilenet_v2.preprocess_input(x), y)
)
validation_dataset = validation_dataset.map(
    lambda x, y: (tf.keras.applications.mobilenet_v2.preprocess_input(x), y)
)
```

```python
base_model = tf.keras.applications.MobileNetV2(input_shape=(60, 60, 3), include_top=False, weights='imagenet')
base_model.trainable = False
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(60, 60, 3)),
    base_model,
    tf.keras.layers.GlobalAveragePooling2D(),
    tf.keras.layers.Dense(len(class_names), activation='softmax')
])
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(train_dataset, validation_data=validation_dataset, epochs=10)
model.save('clasificacion.keras')
```
